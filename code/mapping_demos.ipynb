```python
import pandas as pd

df = pd.read_csv(r'//Users/gavin/Documents/Metis/Coursework/Week_1/Project_1/reference_data/coords_fix.csv')
df.head(15)
```

```python

large=df['Volume'].max()
small=df['Volume'].min()

rad=5
```

```python
import folium


folium_map = folium.Map(location=[40.738, -73.98],
                        zoom_start=11,
                        tiles="CartoDB positron")
folium_map.save('new_york.html')



```

```python
#women

for index, row in df.iterrows():
    long =row['Long']
    lat =row['Lat']
    color = row['Women']
    popup_text = """{}"""
    popup_text = popup_text.format(row['Station'])
    marker = folium.CircleMarker(location = [lat, long], radius=rad*(row['Volume']/small), color=color, fill=True, popup=popup_text)
    
    marker.add_to(folium_map)
    
folium_map.save('new_york_women.html')
```
